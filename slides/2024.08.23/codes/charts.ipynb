```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd

# Data for the models, prices, and launch dates
fn = '/home/michael/github/formace/slides/2024.08.23/codes/openai-embedding-prices.csv'
prices = pd.read_csv(fn)

# Custom function to format the x-axis ticks to display '0.001', '0.0001', etc.
def custom_formatter_exp(x, pos):
    if x < 1:
        return f'{x:.6f}'.rstrip('0').rstrip('.')
    else:
        return f'{x:.1f}'

# Create a horizontal bar chart with seaborn
plt.figure(figsize=(7, 6))
# x=prices[Model], y=prices['Price Value'']
sns.barplot(x=prices['Price Value'], y=prices['Nickname'], palette='viridis')

# Set the x-axis to log scale with custom formatting for small numbers
plt.xscale('log')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(custom_formatter_exp))
plt.gca().xaxis.set_minor_formatter(ticker.NullFormatter())
plt.gca().tick_params(axis='x', rotation=45)

# Add labels and title
plt.xlabel('Price per 1M Tokens (USD)')
plt.ylabel('Embedding Models')
plt.title('Price Comparison of OpenAI Embedding Models')

# Display the plot
plt.show()
```
